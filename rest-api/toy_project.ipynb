* 데이터를 부풀리기 위해 ImageDataGenerator 를 이용하였습니다.
* ImageDataGenerator는 회전 혹은 가로/세로 이동, 확대 등의 무작위 변형을 수행합니다.

```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import matplotlib.pyplot as plt
%matplotlib inline

datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')

img = load_img('data/three/num3.jpeg')  # this is a PIL image
x = img_to_array(img)  # this is a Numpy array with shape (480, 360, 3)
x = x.reshape((1,) + x.shape)  # this is a Numpy array with shape (1, 480, 360, 3)

# the .flow() command below generates batches of randomly transformed images
# and saves the results to the `preview/` directory
plt.figure(figsize=(15,10))
plt.suptitle("Example of image data generator's output")    
i=0
for batch in datagen.flow(x, batch_size=1, save_to_dir='preview', save_prefix='three', save_format='jpeg'):
    i += 1
    if i>20 :
        break
    plt.subplot(4,5,i)
    plt.imshow(batch.reshape(batch.shape[1],batch.shape[2],batch.shape[3]) / 255)
plt.show()

```
